# Count occurrences of "#Omya#" in the timestamp column

```python
num_omya = df[df['Timestamp'] == "#Omya#"].shape[0]

print(f"Number of occurrences of '#Omya#' in Timestamp column: {num_omya}")
```

#### Identify rows where Timestamp is "#Omya#"
#### Replace "#Omya#" with correct timestamps by adding 1 hour to the previous valid timestamp
#### Convert the entire column to datetime format after correction

```python
df4 = df.copy()

omya_rows = df4[df4['Timestamp'] == "#Omya#"].index

for idx in omya_rows:
    df4.at[idx, 'Timestamp'] = pd.to_datetime(df4.at[idx - 1, 'Timestamp']) + pd.Timedelta(hours=1)

df4['Timestamp'] = pd.to_datetime(df4['Timestamp'])
```

```python
df4.to_csv('../Data/df4.csv', index=False)
```
