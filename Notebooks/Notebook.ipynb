# Count occurrences of "#Omya#" in the timestamp column

```python
num_omya = df[df['Timestamp'] == "#Omya#"].shape[0]

print(f"Number of occurrences of '#Omya#' in Timestamp column: {num_omya}")
```

#### Identify rows where Timestamp is "#Omya#"
#### Replace "#Omya#" with correct timestamps by adding 1 hour to the previous valid timestamp
#### Convert the entire column to datetime format after correction

```python
omya_rows = df[df['Timestamp'] == "#Omya#"].index


for idx in omya_rows:
    df.at[idx, 'Timestamp'] = pd.to_datetime(df.at[idx - 1, 'Timestamp']) + pd.Timedelta(hours=1)

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

#### Check if "#Omya#" still exists

```python

if "#Omya#" in df['Timestamp'].values:
    print("There are still occurrences of '#Omya#'. Further investigation needed.")
else:
    print("All '#Omya#' values have been successfully replaced.")
```

### Identify numeric columns stored as objects and convert them to float if needed handling errors

```python
numeric_cols = ["Plant Load (PL)", "Generation behind the Meter (GBM)", "Net Plant Load (NPL)", 
                "Hedged Base Band (HBB)", "Hedged Base Band Cost (HBC)", "Power Market Price (SPOT)", 
                "Additional Power Price (Add.)", "SPOT + Add.Price"]
```
